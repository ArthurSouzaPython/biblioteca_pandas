```python
import pandas as pd
import numpy as np

df = pd.read_csv("ks-projects-201801.csv", sep = ",")
df.head()
```

```python
df.info()
```

```python
# df = df.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
df = df.drop(columns=['name', 'category', 'goal', 'pledged', 'usd pledged'])
df.head()
```

```python
df.dtypes
```

```python
df['usd_goal_real'] = df['usd_goal_real'].astype(str)
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
df.head()
```

```python
df['launched'].replace('12:12', ' ')
```

```python
df['launched'] = df['launched'].replace(' \d\d:\d\d:\d\d', '', regex=True)
# \d é um digito , tira o horario do campo data
```

```python
df.head()
```

```python
df['usd_goal_real'] = df['usd_goal_real'].astype('float').astype('int64')
# Muda Type da df['usd_goal_real']
```

```python
df.dtypes
```

```python
df['deadline'] = pd.to_datetime(df['deadline'], format='%Y-%m-%d')
df['launched'] = pd.to_datetime(df['launched'], format='%Y-%m-%d')

# Transforma em tipo data com Pandas
```
